```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from tensorflow.keras.utils import to_categorical
```

```python
dataset = pd.read_csv('../dataset/iris.csv')
```

```python
dataset.head()
```

```python
drop_data = ['species']
dataset = dataset.drop(drop_data, axis =1)
```

```python
print(dataset.shape)
dataset.head()
```

```python
X = dataset.iloc[:, :3].values  # Mengambil 3 kolom pertama sebagai atribut
y = dataset.iloc[:, 3].values  # Mengambil kolom keempat sebagai label
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(16, input_shape=(3,), activation='relu')) #Input Layer
model.add(tf.keras.layers.Dense(16, activation='relu')) #Hidden Layer 1
model.add(tf.keras.layers.Dense(16, activation='relu')) #Hidden Layer 2
model.add(tf.keras.layers.Dense(16, activation='relu')) #Hidden Layer 3
model.add(tf.keras.layers.Dense(y.shape[1], activation='softmax'))
```

```python
# Compile model dengan optimizer Adam
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=300, batch_size=32)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
```

```python
print('Loss:', loss)
print('Accuracy:', accuracy)
```
